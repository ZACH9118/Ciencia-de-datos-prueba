```python
df.rename(columns={"Año":"Fecha"},inplace=True)
#df.info()
df['Fecha'] = df['Fecha'].str.replace(',','')

df['Cantidad de empleos'] = df['Cantidad de empleos'].str.replace(',','')
df['Activos (Miles)'] = df['Activos (Miles)'].str.replace(',','')
df['Cantidad'] = df['Cantidad'].str.replace(',','')
#print(df.head())
```

```python
# errors='coerce' convierte los datos erróneos en NaT (Not a Time) en lugar de dar error
#df['Fecha'] = pd.to_datetime(df['Fecha'], errors='coerce')
#df.head()
df["Fecha"] = df["Fecha"].astype(int)
df["Cantidad de empleos"] = df["Cantidad de empleos"].astype(int)
df["Activos (Miles)"] = df["Activos (Miles)"].astype(int)
df["Cantidad"] = df["Cantidad"].astype(int)

df.info()
#print(df.head())
```
